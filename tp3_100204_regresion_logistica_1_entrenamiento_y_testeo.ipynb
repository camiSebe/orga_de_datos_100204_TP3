### Búsqueda de hiperparámetros

A partir de los resultados obtenidos en el otro archivo:

```python
best_c = 2.195254015709299
best_penalty = 'l1'
```

```python
def calcular_regresion(mejor_c, mejor_penalty, random_inicial):
    X_train, X_test, y_train, y_test = train_test_split(points, labels, test_size = 0.5, random_state = random_inicial)
    X_train.shape
    X_test.shape
    model = LogisticRegression(penalty = mejor_penalty, C=best_c, solver='saga')
    model.fit(X_train, y_train)
    print(classification_report(y_train, model.predict(X_train)))
    roc_y_train = roc_auc_score(y_train, model.predict_proba(X_train)[:,1])
    print(classification_report(y_test, model.predict(X_test)))
    roc_y_test = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
    return(roc_y_train, roc_y_test)
```

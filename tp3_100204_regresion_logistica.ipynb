```python
def calcular_regresion(random_inicial):
    X_train, X_test, y_train, y_test = train_test_split(points, labels, test_size = 0.5, random_state = random_inicial)
    X_train.shape
    X_test.shape
    model = LogisticRegression(C=0.5)
    model.fit(X_train, y_train)
    print(classification_report(y_train, model.predict(X_train)))
    roc_y_train = roc_auc_score(y_train, model.predict_proba(X_train)[:,1])
    print(classification_report(y_test, model.predict(X_test)))
    roc_y_test = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
    return(roc_y_train, roc_y_test)
```

### ¿Cuál es el peor resultado que puedo obtener?

```python
peor_random = 0
peor_roc1 = 0.7125194003096725
peor_roc2 = 0.7061256485253561

for i in range(100):
    print( "*****", i, "*****")
    X_train, X_test, y_train, y_test = train_test_split(points, labels, test_size = 0.5, random_state = i)
    model = LogisticRegression(C=0.5)
    model.fit(X_train, y_train)
    # print(classification_report(y_train, model.predict(X_train)))
    roc1_actual = roc_auc_score(y_train, model.predict_proba(X_train)[:,1])
    # print(classification_report(y_test, model.predict(X_test)))
    roc2_actual = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
    
    if (roc1_actual < peor_roc1):
        print("EMPEORO")
        peor_random = i
        peor_roc1 = roc1_actual
        peor_roc2 = roc2_actual
        print("peor_random = ", peor_random, "peor_roc1 = ", peor_roc1, "peor_roc2 = ", peor_roc2)
```

```python
peor_resultado_y_train, peor_resultado_y_test = calcular_regresion(peor_random)
print("peor_resultado_y_train = ", peor_resultado_y_train)
print("peor_resultado_y_test = ", peor_resultado_y_test)
```

```python
print("peor_random = ", peor_random)
```
